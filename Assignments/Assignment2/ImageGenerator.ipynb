```python
import os
import matplotlib.pyplot as plt

from PIL import Image
```

```python
def resize_image(image_path, output_size=(224, 224)):
    image = Image.open(image_path)
    width, height = image.size
    new_size = min(width, height)
    
    left = (width - new_size) / 2
    top = (height - new_size) / 2
    right = (width + new_size) / 2
    bottom = (height + new_size) / 2
    
    image_cropped = image.crop((left, top, right, bottom))
    
    image_resized = image_cropped.resize(output_size)
    image_resized = image_resized.rotate(270)
    
    return image_resized
```

```python
StudentNumber = "101082714" # Your student number
IMGPATH = "/Users/seonghak/Downloads/Assignment2/Ass2" # original Image Path
RESIZEIMGPATH = "Individual" #Save Path 

img_list = os.listdir(IMGPATH)
img_list.sort()
img_list = img_list[1:]
```

```python
# Check crop and resize image (No cutting your finger!)

fig, ax = plt.subplots(9, 3, figsize=(3, 8))

for i, img in enumerate(img_list):
    img_new = resize_image(os.path.join(IMGPATH, img))
    img_new = img_new.rotate(90)
    ax[i//3, i%3].imshow(img_new)
    ax[i//3, i%3].set_xticks([])
    ax[i//3, i%3].set_yticks([])
plt.show()
```

```python
alphabet = ["A", "B", "C", "D", "E", "F", "G", "H", "I"]
pos = 0
idx = 1
for i, img in enumerate(img_list):
        
    img_new = resize_image(os.path.join(IMGPATH, img))
    
    if not os.path.exists(RESIZEIMGPATH):
        os.makedirs(RESIZEIMGPATH)
        
    img_new.save(os.path.join(RESIZEIMGPATH, f"{StudentNumber}_{alphabet[pos]}_{idx}.jpg"))
    
    idx += 1
    
    if (i+1) % 3 == 0:
        pos += 1
        idx = 1
```
